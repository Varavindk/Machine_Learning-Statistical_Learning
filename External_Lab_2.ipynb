### Find out the mean, median and variance for the column - DailyRate 

```python
median=0

#calculate mean from desc 
mean=dataset['DailyRate'].describe()[1]

#Fetch std from desc
std=dataset['DailyRate'].describe()[2]

#calculate median from desc
length = len(dataset)
if(length%2==0):
    midlength=int(length/2)
    middata1=dataset['DailyRate'] [midlength]
    midlength=midlength+1
    middata2=dataset['DailyRate'] [midlength]
    median=round(((middata1)+(middata2))/2) 
else:
    midlength=int((length/2))
    median=dataset['DailyRate'] [midlength]

print( "Daily Rate mean = {}".format(str(mean)))
print( "Daily Rate Median = {}".format(str(median)))
print( "Daily Rate variance = {}".format(str(std*std)))


```
